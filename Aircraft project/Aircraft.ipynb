```python
Flight_type_Landing_counts = pd.crosstab([df['Operating Airline']], 
                         [df['Flight Type']], 
                         values=df['Landing Count'], 
                         aggfunc='count')
Flight_type_Landing_counts = Flight_type_Landing_counts.fillna(0)
Flight_type_Landing_counts['Total Landings'] = Flight_type_Landing_counts['Domestic'] + Flight_type_Landing_counts['International']
Flight_type_Landing_counts.sort_values(by = 'Total Landings',ascending = False)
```
